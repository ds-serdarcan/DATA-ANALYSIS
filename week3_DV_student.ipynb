Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

__2. Make a bar plot__

__matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)__<br><br>

__https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.bar.html__<br><br>

__https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html__

```python
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
x = np.arange(len(index)) # the label locations
width = 0.35 # the width of the bars
fig, ax = plt.subplots(figsize=(7, 5))
scores1 = ax.bar(x - width / 2, df.income, width, label="income")    
scores2 = ax.bar(x + width / 2, df.expense, width, label="expense") 

# We may add text for labels, title and custom x-axis tick labels 
# ax.set_ylabel("Scores")
# ax.set_title("Scores of Income and Expense")
ax.set_xticks(x)
ax.set_xticklabels(index, rotation=90)
# ax.bar_label(scores1, padding=3)
# ax.bar_label(scores2, padding=3)
ax.legend()
fig.tight_layout()


```

__3. Stack the bars__

```python
fig, ax = plt.subplots(figsize=(5.5, 3.5))

width = 0.55  # the width of the bars: can also be len(df.index) sequence
ax.bar(df.index, df.income, width, label="income")    
ax.bar(df.index, df.expense, width, bottom=df.income, label="expense") 
ax.set_xticks(x)
ax.set_xticklabels(index, rotation=90)
ax.legend()
fig.tight_layout()
```

__4. Rotate the labels and set figsize__

```python
fig, ax = plt.subplots(figsize=(7.5, 5))

width = 0.45  
ax.bar(df.index, df.income, width, label="income")    
ax.bar(df.index, df.expense, width, bottom=df.income, label="expense") 
ax.set_xticks(x)
ax.set_xticklabels(index, rotation=45)
ax.legend()
fig.tight_layout()
```

__5. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"__<br><br>


```python
fig, ax = plt.subplots()

width = 0.45  
ax.bar(df.index, df.income, width, label="income")    
ax.bar(df.index, df.expense, width, bottom=df.income, label="expense") 
ax.set_xticks(x)
ax.set_xticklabels(index)
ax.legend()

new_index = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
ax.set_xticklabels(new_index)

fig.tight_layout()
```

__6. Unstuck the bars, annotate the hight of the bars on top of them__

```python
fig, ax = plt.subplots(figsize=(7.5, 5.5))

width = 0.25  
scores1 = ax.bar(x - width / 2, df.income, width, label="income")    
scores2 = ax.bar(x + width / 2, df.expense, width, label="expense") 
ax.set_xticks(x)
ax.set_xticklabels(index)
plt.legend()

new_index = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
ax.set_xticklabels(new_index)
ax.bar_label(scores1, padding=3)
ax.bar_label(scores2, padding=3)

fig.tight_layout()
```

__8. Plot the avg tip by size  (lineplot)__

```python
tip_plot = tips.groupby("size")[["tip"]].mean()
tip_plot
```

```python
size = tips.groupby("size")[["tip"]].mean().index
size
```

```python
fig, ax = plt.subplots()

ax.plot(size, tip_plot)
ax.set_xlabel("size")

plt.show()
```

__9.Set the linestyle as "--"__

```python
fig, ax = plt.subplots()

ax.plot(size, tip_plot,lw=1.5, linestyle="--") # dashes
ax.set_xlabel("size")    

plt.show()
```
