# Distribution Plots
## rugplot, displot,histplot,kdeplot,dimensional kde plot

```python
plt.figure(figsize=(14,9))
sns.scatterplot(data=df, x="salary", y="sales")
sns.rugplot(x="salary", y="sales", data=df, color="red")
```

```python
sns.displot(data=df, x="salary",kde=True,bins=50)
```

```python
sns.kdeplot(data=df, x="salary",bw=0.1)
```

```python
sns.histplot(data=df, x="salary",kde=True,bins=100)
sns.rugplot(data=df, x="salary",color="red",height=0.1)
```

```python
sns.displot(data=df, x="salary",kde=True,bins=100)
```

```python
sns.histplot(data=df, x="salary",y="sales",bins=100,kde=True)
```

```python
plt.figure(figsize=(15,8),dpi=300)
sns.displot(data=df, x="salary",y="sales",bins=100,color="blue")
sns.kdeplot(data=df, x="salary",y="sales",color="yellow",alpha=0.5)
```

```python
sns.displot(data=df, x="salary",y="sales",bins=10,color="red",
            lw=3,ls="-.",
            legend=True
           )
```

```python
sns.displot(data=df, x="salary",y="sales",bins=10,color="red",
            lw=3,ls="-.",
           )
```

```python
sns.histplot(data=df, x="salary",bins=10,stat="probability"
             ,fill=False,element="step",cumulative=True
           )  # ECDF
```

```python
sns.histplot(data=df, x="salary",hue="division",element="poly"
           )
```

```python
sns.histplot(data=df, x="salary",hue="division",multiple="stack"
           )
```

```python
sns.histplot(data=df, x="salary",hue="division",multiple="fill"
           )
```

```python
sns.histplot(data=df, x="salary",y="sales", hue="division",bins=30
           )
```

```python
sns.histplot(data=df, x="salary",y="division", hue="division",bins=200
           )
```

```python
ax = sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="work experience", 
             hue="division",multiple="dodge",shrink=0.5,bins=5)

sns.move_legend(obj=ax, loc="center", bbox_to_anchor=(1.2, 0.6),title="Serdar", frameon=True)

for i in range(len(ax.containers)):
    ax.bar_label(ax.containers[i],label_type="edge",padding=0.2)

plt.show()
```
